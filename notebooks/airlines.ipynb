```python
import forcebundle.readDataUtils as rd
import forcebundle.ForcedirectedEdgeBundling as feb
import math
import statistics as st
```

# Caso de ejempo: Aerol√≠neas de EEUU

Este es el caso de ejemplo dado por el repositorio [d3.ForceBundle](https://github.com/upphiminn/d3.ForceBundle).

```python
transf_function = lambda row: (float(row[0]), -float(row[1]), float(row[2]), -float(row[3]))

csvfile = '../data/airlines.csv'
edges, bounds = rd.read_edges_from_csv(csvfile, transf_function)
```

```python
import forcebundle.usageHelper as uh
input_lines = feb.edges2lines(edges)

uh.plot_lines(input_lines, margin = 0.1)
```

```python
feb.S_initial = st.median([feb.edge_length(edge) for edge in edges])/100
output_lines = feb.forcebundle(edges)
```

```python
uh.plot_lines(output_lines, margin = 0.1)
```
